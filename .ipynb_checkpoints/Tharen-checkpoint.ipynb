It is desired to generate random numbers from the continuous distribution function given by,

$$f(x) = sin(π x)$$

# Task 1:
## Develop an algorithm based on the analytical inversion technique to generate random numbers from $f(x)$. 
### Add your answer in the Markdown cell below

1. Obtain the Cumulative Distributive Function (CDF)
2. Obtain the inverse of the CDF
3. Assume a random number, $U$, from a uniform distribution.
4. Pass $U$ as a parameter to the inverse of the CDF, from which a random number is generated.

# Task 2: 
## Code the algorithm in Python

```python
# Check that distribution is proper
from scipy.integrate import quad

def integrand(x, c = 1):
    return c * np.sin(np.pi * x)

lower, upper = 0, 1

area, err = quad(integrand, lower, upper)

## Cons is the constant applied to the integrand to make the distribution proper
cons = area**-1

# Obtain CDF for Proper Distribution
## Integral of cons * f(x) = -Cos[Pi x]/2
def integral(x):
    return -np.cos(np.pi * x)/2

# Obtain Inverse CDF for Proper Distribution
## Inverse of Integral = Integral^(-1) = ±(ArcCos[1 - 2x]/Pi)
def inv_integral(x):
    return np.arccos(1 - 2*x)/np.pi

# Generete Random Number using Inverse CDF
def randAI():
    return inv_integral(np.random.rand())

print ("Random Number Generated using Analytical Inversion:")
print (randAI())
```

# Task 3:
## Using your mathematical algorithm generate 1000 numbers
## Plot a histogram that represents the distribution (use 11 bins) 
## Obtain a goodness of fit statistic for the distribution
## Plot a two variable scatter-plot to demonstrate that the numbers are not correlated 

```python
# Generate 1000 Numbers
def generate(algorithm, amount = 1000):
    array = []
    for _ in range(amount):
        array.append(algorithm())
    #return array
    return np.array(array)

arand_AI = generate(randAI)

# Plot a Histogram
obs_AI, bins_obs_AI = np.histogram(arand_AI, bins=11)
width_AI = 0.7 * (bins_obs_AI[1] - bins_obs_AI[0])
center_AI = (bins_obs_AI[:-1] + bins_obs_AI[1:]) / 2
plt.bar(center_AI, obs_AI, align='center', width=width_AI)
plt.xlabel("Number Generated")
plt.ylabel("Frequency")
plt.title("Histogram of Frequency for Random Numbers Generated")
plt.show()

aexp_AI = np.array(inv_integral(np.arange(0, 1, 1/1000.)))
exp_AI, bins_exp_AI = np.histogram(aexp_AI, bins=11)

# Obtain Goodness of Fit
chi_2_obt_AI = np.sum((obs_AI-exp_AI) ** 2 / exp_AI)
print ("chi_2_obt_AI= ", chi_2_obt_AI)
#
chi_2_AI=stats.chi2.ppf(q=1.0-.05,df=10)
print ("chi_2 from table = ", chi_2_AI)

if chi_2_obt_AI > chi_2_AI:
    print ("We reject the hypothesis that the two distributions are similar with a confidence of 95%")
else:
    print ("We assume that the distributions are similar with a confidence of 95%")
    
# Check Variables are not correlated
for i in range(len(arand_AI)):
    plt.scatter(arand_AI[i], aexp_AI[i]) 
plt.xlabel("Observed Numbers")
plt.ylabel("Expected Numbers")
plt.title("Scatter Plot between Observed and Expected Random Numbers Generated")
plt.show()

print ("From the Scatter Plot, it has been clearly shown that there is no correlation.")
```

# Task 4:
## Write down  an algorithm (in pseudo code) based on the accept-reject method to generate random numbers described by $f(x)$. 
### Add your answer below in the Markdown cell below

1. Assume $f(x)$ is the required distribution.
2. Define a function $h(x) = M$, where $h(x) > f(x)$ for the specified range.
3. Generate $U$ in the given range.
4. Generate $V$ in the range $[0, M]$
5. If $V < f(U)$, accept $U$; else repeat from step 3.

# Task 5:
## Code the algorithm in Python

```python
# Define M
## From observing a plot of integrand(x, cons), 1.25 would be sufficiently large.

# Generate an acceptable U
def randARM(M = 1.25):
    while True:
        U = np.random.rand()
        V = np.random.rand() * M
        if  V < integrand(U, cons): # cons defined in Task 2, to achieve a proper distribution
            return U

print ("Random Number Generated using Accept-Reject Method:")
print (randARM())
```

# Task 6
## Using your accept-reject algorithm generate 1000 numbers
## Plot a histogram that represents the distribution (use 11 bins) 
## Obtain a goodness of fit statistic for the distribution

```python
# Generate 1000 Numbers
arand_ARM = generate(randARM)

# Plot a Histogram
obs_ARM, bins_obs_ARM = np.histogram(arand_ARM, bins=11)
width_ARM = 0.7 * (bins_obs_ARM[1] - bins_obs_ARM[0])
center_ARM = (bins_obs_ARM[:-1] + bins_obs_ARM[1:]) / 2
plt.bar(center_ARM, obs_ARM, align='center', width=width_ARM)
plt.xlabel("Number Generated")
plt.ylabel("Frequency")
plt.title("Histogram of Frequency for Random Numbers Generated")
plt.show()

aexp_ARM = np.array(integrand(np.arange(0, 1, 1/1000.), cons))
exp_ARM, bins_exp_ARM = np.histogram(aexp_ARM, bins=11)

# Obtain Goodness of Fit
chi_2_obt_ARM = np.sum((exp_ARM-obs_ARM) ** 2 / exp_ARM)
print ("chi_2_obt_ARM = ", chi_2_obt_ARM)
#
chi_2_ARM = stats.chi2.ppf(q=1.0-.05,df=10)
print ("chi_2 from table = ", chi_2_ARM)

if chi_2_obt_ARM > chi_2_ARM:
    print ("We reject the hypothesis that the two distributions are similar with a confidence of 95%")
else:
    print ("We assume that the distributions are similar with a confidence of 95%")
```

## Experimentally measure the computational time for both algorithms and test for any significant difference

```python
# Test Computational Time
import time as t

def compute_times(algorithm, times = 100, ins = 10000):
    array_time = []
    for _ in range(times):
        start = t.time()
        for _ in range(ins):
            algorithm()
        stop = t.time()
        array_time.append(float(stop-start) / float(times))
    return np.array(array_time)

time_AI = compute_times(randAI)
time_ARM = compute_times(randARM)

if(np.average(time_AI) < np.average(time_ARM)):
    faster = "AI"
else:
    faster = "ARM"

print("Time comparison between AI and ARM respectively:")
print("AI: ", np.average(time_AI), "\tARM: ", np.average(time_ARM))
print(faster, "is faster")
```

```python
#One Way Anova for Comparison 
sig_value = 0.05
F, p_val = stats.f_oneway(time_AI, time_ARM)

if(p_val <= sig_value):
    print ("There is a significant difference, with", str.format('{0:.2f}', F),"variation between samples")
else:
    print ("There is no siginificant difference")
```
